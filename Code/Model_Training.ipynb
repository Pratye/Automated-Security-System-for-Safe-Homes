# Reffer to make the train.txt and test.txt

```python
import os


for i in os.listdir('/content/drive/MyDrive/Colab/Security Surveillance/darknet/build/darknet/x64/data/obj'):
  if i=='README.roboflow.txt' or i=='roboflow.zip' or i=='train1' or i=='test1'or i=='train':
    continue
  else:
    x = os.path.join('/content/drive/MyDrive/Colab/Security Surveillance/darknet/build/darknet/x64/data/obj',i)
    a = os.path.join('build/darknet/x64/data/obj', i)
    f = open('/content/drive/MyDrive/Colab/Security Surveillance/darknet/build/darknet/x64/data/test.txt', 'w+')
    for img in os.listdir(x):
      if img=='_darknet.labels' or img.endswith('txt')==True:
        continue
      else:
        #print(os.path.join(a, img))
        f.write(os.path.join(a, img)+'\n')

        #/content/darknet/cfg/yolov4-custom-detector.cfg
```
